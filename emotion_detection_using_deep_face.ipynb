**Design an AI-enabled application for emotion/gesture detection.**

Deepface is a lightweight face recognition and facial attribute analysis (age, gender, emotion and race) framework for python. It is a hybrid face recognition framework wrapping state-of-the-art models: VGG-Face, Google FaceNet, OpenFace, Facebook DeepFace, DeepID, ArcFace and Dlib. Those models already reached and passed the human level accuracy. The library is mainly based on Keras and TensorFlow.

```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt

img1 = cv2.imread("/content/image 1.webp")

if img1 is None:
    print("Image not found. Please check the path.")
else:
    img1_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
    plt.imshow(img1_rgb)
    plt.axis('off')
    plt.show()
```

```python
# Plotting the image
# Flipping the color from BGR to RGB
plt.imshow(img1[:, :, ::-1 ])
plt.show()
```

```python
##Analyze different features of the Image
obj=DeepFace.analyze(img1, actions = ['age', 'gender', 'race', 'emotion'])
print(obj)
```

```python
##Print Result
print("Your age is",obj[0]['age'],"and you are",obj[0]["dominant_race"],"and current emotion is ",obj[0]["dominant_emotion"],"and your gender capture is", obj[0]['dominant_gender'])
```
